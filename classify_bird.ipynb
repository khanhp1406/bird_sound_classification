```python
import pandas as pd
df = pd.read_csv('bird_audio.csv')
```

```python
df
```

```python
from collections import defaultdict
import ast

data = defaultdict(list)

for idx, row in df.iterrows():
    array = []
    feature = row['feature']
    for idx, num in enumerate(feature[1:-1].split()):
        data[f'feature_{idx}'].append(float(num))
    data['class_number'].append(row['class_number'])
data = pd.DataFrame(data)
data
```

```python
y = data['class_number']
y
```

```python
X = data.drop(['class_number'], axis=1)
X
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test , y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score

svc = SVC(kernel='linear', gamma=0.1)
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
train_accuracy = svc.score(X_train, y_train)
test_accuracy = svc.score(X_test, y_test)
```

```python
print(train_accuracy)
print(test_accuracy)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print("Accuracy: ", accuracy_score(y_test, svc.predict(X_test)))
```

```python
#save model
import pickle

pickle.dump(svc, open('svc_bird.pkl', 'wb'))
```

```python
import librosa, librosa.display
import pandas as pd
import sklearn
import numpy as np
```

```python
with open('svc_bird.pkl', 'rb') as file:  
    model = pickle.load(file)
```

```python
SR = 44100           # Sampling frequency
N_MELS = 40          # Mel band parameters   

def extract_features(file_name):
    try:
        file_name = file_name.replace('\\', '/')
        audio, sample_rate = librosa.load(file_name, mono=True, sr=SR, res_type='kaiser_fast')
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=N_MELS)
        mfccs_scaled = np.mean(mfccs.T, axis=0)
    except Exception as e:
        print(f'Error encountered while parsing file: {file_name}')
        return None
    return mfccs_scaled
```
